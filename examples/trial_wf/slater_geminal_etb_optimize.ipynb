```python
import os
import h5py
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
alpha_s = 0.2
alpha_p = 0.2

df = []
ccvtz_energy = -16.8069
for n_s in range(0, 6):
    for n_p in range(6):
        for n_d in range(0, 4):
            hdf_file = f"data/slater_geminal_etb_{n_s}_{alpha_s}_{n_p}_{alpha_p}_{n_d}.hdf5"
            if os.path.exists(hdf_file):
                print(f"File {hdf_file} exists")
                with h5py.File(hdf_file, "r") as f:
                    energy = f["energy"][()][-1]
                    error = f['energy_error'][()][-1]
                    if 'time' in f.keys():
                        time = f["time"][()]
                    else:
                        time = 0
                df.append({
                    "n_s": n_s,
                    "n_p": n_p,
                    "n_d":n_d,
                    "energy": energy,
                    'error': error,
                    "time": time,
                })


df = pd.DataFrame(df)
df
```

```python
g = sns.relplot(data =df, x="n_s", y="energy", hue="n_p", col="n_d")
for ax in g.axes.flat:
    ax.axhline(ccvtz_energy, linestyle="--", color="k", label="ccvtz energy")
```

```python
ax = sns.lineplot(data =df, x="n_s", y="time", hue="n_p", marker="o")
```
